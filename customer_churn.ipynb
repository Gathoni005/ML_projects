SPLITTING OF THE DATASET


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

the code snippet from scikit learn library for logistic regression

```python
x=customer_churn_df.drop(columns=["Churn"])
y=customer_churn_df["Churn"]
print(f"Features (x) shape: {x.shape}")
print(f"Target (y) shape: {y.shape}")
```

features set x contains 7043 rows and 19columns
target variable y contains 7043 entries

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

splitting the datasets into training and testing sets

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
```

approximately 80% of all predictions was correct
